### **Creating Distributions from Data**

In data analysis, understanding how variables vary is essential for making informed decisions. Distributions help engineers and analysts identify patterns, detect anomalies, and model real-world systems. A random variable represents a quantity whose values are uncertain, while realizations are the past observed values collected through sampling. Since it is rarely practical to analyze an entire population, analysts work with representative samples to generalize findings. Visualization techniques like histograms, frequency polygons, and relative frequency plots are critical tools for revealing the structure and variability of data.

#### **Core Concepts**
1. Random Variables and Variation
    - Random variables represent uncertain values affected by external factors.
    - Variation occurs when a variable’s values differ across observations, such as over time or across customers.
    - Descriptive analytics uses data visualization to understand these variations and their impact.
2. Types of Data
    -.Categorical Data: Consists of labels or names that cannot be mathematically manipulated (e.g., beverage types, product categories).
    - Quantitative Data: Numeric values that can be mathematically analyzed (e.g., sales numbers, temperatures).
3. Frequency Distributions
    - Show how often specific values or ranges of values occur.
    - Categorical Example: Soft drink sales distribution to compare popularity.
    - Quantitative Example: Age distribution using ranges (bins) to capture continuous data.
4. Relative and Percent Frequency
    - Relative frequency expresses counts as proportions of the total.
    - Percent frequency provides percentages, useful for estimating probabilities.
    - Example: Using purchase data to predict required stock for a concert.
5. Benford’s Law
    - Observes that in many natural datasets, lower digits (like 1) appear more frequently as the first digit.
    - Used in fraud detection and anomaly detection, especially in financial datasets.

#### **Visualization Techniques**
1. Histograms
    - Best for quantitative data.
    - Columns without spaces represent continuous ranges of values.
    - The number of bins and bin width strongly affect clarity:
    - Too many bins make data noisy.
    - Too few bins oversimplify patterns.
2. Frequency Polygons
    - Use lines instead of bars, allowing easier comparison between multiple variables.
    - Example: Comparing age distributions of male and female individuals.
3. Skewness
    - Left-skewed: Most values are high, few are low (e.g., exam scores).
    - Right-skewed: Most values are low, few are high (e.g., income data).
    - Symmetric: Balanced distribution, like heights or standardized test scores.

#### **Key Steps for Building Distributions**
1. Define Bins: Decide the number and range of bins (typically 5–20).
2. Calculate Frequencies: Use software functions like COUNTIF or FREQUENCY in Excel.
3. Visualize:
    - Clustered column charts for categorical data.
    - Histograms or frequency polygons for quantitative data.
4. Interpret Patterns: Identify trends, anomalies, or potential risks.

```python
# Setup
# ========================================
# Generate sample quantitative data
# ========================================
# Simulated data: 200 random ages between 15 and 70
ages = np.random.randint(15, 70, 200)

# Define bin ranges for age groups
bins = [15, 25, 35, 45, 55, 65, 75]

# ========================================
# Part 1: Core Concepts
# ========================================
# Calculate frequency and relative frequency
frequency, bin_edges = np.histogram(ages, bins=bins)
relative_frequency = frequency / len(ages)

# Create a frequency table
distribution_df = pd.DataFrame({
    "Age Range": [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)],
    "Frequency": frequency,
    "Relative Frequency": relative_frequency.round(3)
})

print("=== Part 1: Core Concepts ===")
print("Frequency and Relative Frequency Table:")
print(distribution_df)

# ========================================
# Part 2: Visualization Techniques
# ========================================
# Calculate midpoints for the frequency polygon
midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])

plt.figure(figsize=(10, 5))

# Histogram
plt.hist(ages, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')

# Frequency Polygon
plt.plot(midpoints, frequency, color='red', marker='o', linewidth=2, label='Frequency Polygon')

print("=== Part 2: Visualization Techniques ===")
plt.title("Part 2: Visualization Techniques - Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.xticks(bins)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# ========================================
# Part 3: Key Steps for Building Distributions
# ========================================

print("\n=== Part 3: Key Steps for Building Distributions ===")

# Step 1: Define bins
print(f"Step 1: Bins defined as {bins}")

# Step 2: Calculate frequency and relative frequency
frequency, bin_edges = np.histogram(ages, bins=bins)
relative_frequency = frequency / len(ages)

step2_df = pd.DataFrame({
    "Age Range": [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)],
    "Frequency": frequency,
    "Relative Frequency": relative_frequency.round(3)
})

print("\nStep 2: Frequency Table")
print(step2_df)

# Step 3: Visualize the data
plt.figure(figsize=(10, 5))
plt.hist(ages, bins=bins, color='lightgreen', edgecolor='black', alpha=0.7, label='Histogram')
plt.plot(midpoints, frequency, color='blue', marker='o', linewidth=2, label='Frequency Polygon')

plt.title("Step 3: Visualization of Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.xticks(bins)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Step 4: Interpret Patterns
max_freq_index = np.argmax(frequency)
most_common_range = step2_df.iloc[max_freq_index]["Age Range"]

print("\nStep 4: Interpretation")
print(f"- The most common age range is {most_common_range} with {frequency[max_freq_index]} individuals.")

# Check skewness
lower_half = sum(frequency[:len(frequency)//2])
upper_half = sum(frequency[len(frequency)//2:])

if lower_half < upper_half:
    print("- The data is LEFT-SKEWED (more older individuals).")
elif lower_half > upper_half:
    print("- The data is RIGHT-SKEWED (more younger individuals).")
else:
    print("- The data is approximately symmetric.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Set random seed for reproducibility (same ages every run)
np.random.seed(42)

# Simulated data: 200 random ages between 15 and 70 (fixed now)
ages = np.random.randint(15, 70, 200)

# Define multiple bin configurations to test
# Original: Moderate bins (6 intervals)
bins_original = [15, 25, 35, 45, 55, 65, 75]

# Wide bins: Fewer, broader intervals (4 intervals)
bins_wide = [15, 30, 45, 60, 75]

# Narrow bins: Many, finer intervals (12 intervals)
bins_narrow = [15, 20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75]

# Function to process and visualize for given bins (avoids code duplication)
def process_distribution(bins, title_suffix=""):
    # Calculate frequency and relative frequency
    frequency, bin_edges = np.histogram(ages, bins=bins)
    relative_frequency = frequency / len(ages)

    # Create age range labels
    age_ranges = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]

    # Frequency table
    distribution_df = pd.DataFrame({
        "Age Range": age_ranges,
        "Frequency": frequency,
        "Relative Frequency": relative_frequency.round(3)
    })

    # Calculate midpoints for frequency polygon
    midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])

    # Print header with bin info
    print(f"\n=== Using {len(bins)-1} bins: {bins} {title_suffix}===")
    print("Frequency and Relative Frequency Table:")
    print(distribution_df)

    # Visualization
    plt.figure(figsize=(10, 5))
    plt.hist(ages, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')
    plt.plot(midpoints, frequency, color='red', marker='o', linewidth=2, label='Frequency Polygon')
    plt.title(f"Age Distribution - {title_suffix} Bins")
    plt.xlabel("Age Range")
    plt.ylabel("Frequency")
    plt.xticks(bins)
    plt.legend()
    plt.grid(axis='y', alpha=0.3)
    plt.show()

    # Simple interpretation (most common range)
    if len(frequency) > 0:
        max_freq_index = np.argmax(frequency)
        most_common_range = age_ranges[max_freq_index]
        print(f"Most common age range: {most_common_range} with {frequency[max_freq_index]} individuals.")

# ========================================
# Task 1: Testing Different Bin Sizes
# ========================================

# Test 1: Original (Moderate) Bins
process_distribution(bins_original, "Original (Moderate)")

# Test 2: Wide Bins
process_distribution(bins_wide, "Wide")

# Test 3: Narrow Bins
process_distribution(bins_narrow, "Narrow")
```

### **Statistical Analysis of Distributions of Quantitative Variables**

Understanding data distributions is essential for engineers and analysts working with real-world datasets. Statistical measures and visualization techniques help summarize, analyze, and interpret data patterns effectively. This section focuses on methods for identifying central tendencies, variability, and outliers in quantitative variables. These concepts are foundational for data-driven decision-making, predictive modeling, and system optimization in computer engineering applications.

#### Measures of Location (Central Tendency):
- Identify the "center" of a dataset, which represents a typical value.
- Mean (Average):
    - Sensitive to extreme values or outliers.
    - Useful when data is evenly distributed.
- Median (Middle Value):
    - Preferred when data contains extreme values or is skewed.
    - More robust than the mean for smaller datasets.
- Mode (Most Frequent Value):
    - Best for datasets with a limited range of distinct values.
    - In continuous data, represents the range with the highest frequency in a histogram.

#### Measures of Variability (Dispersion):
- Quantify how spread out the data values are.
- Range:
    - Simple to compute but overly affected by outliers.
- Standard Deviation:
    - Indicates average deviation from the mean.
    - Best for symmetric, bell-shaped distributions.
- Percentiles and Interquartile Range (IQR):
    - Describe data spread without being distorted by extreme values.
    - Useful for skewed datasets.
    - IQR highlights the middle 50% of data values.

#### Visualization Techniques: Box-and-Whisker Plots
- Combine statistical measures into a clear visual summary.
- Show median, quartiles, range, and potential outliers.
- Useful for comparing multiple datasets, such as performance metrics across different systems.
- Outliers are easily identifiable, helping engineers detect anomalies in system behavior or sensor data.

#### Applications in Engineering
- Identifying network latency anomalies using skewness and outlier detection.
- Analyzing processor performance data through measures of variability.
- Monitoring sensor outputs in IoT systems with box plots for quality control.
- Selecting appropriate summary measures for predictive models and machine learning pipelines.


Statistical analysis of quantitative data involves understanding central tendency, variability, and outlier behavior. Combining numerical measures with effective visualizations like box plots enables engineers to interpret data more accurately. By applying these concepts, computer engineers can improve decision-making in areas such as performance optimization, anomaly detection, and predictive analytics.

```python
# ===============================================
# Statistical Analysis of Distributions of Quantitative Variables
# ===============================================

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate sample quantitative data: 200 random ages between 15 and 70
ages = np.random.randint(15, 70, 200)

# -----------------------------------------------
# PART 1: CORE CONCEPTS - Measures of Central Tendency and Variability
# -----------------------------------------------

print("=== PART 1: CORE CONCEPTS ===")

# Calculate measures of central tendency
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = pd.Series(ages).mode()[0]  # mode() returns a Series

# Calculate measures of variability
range_age = np.max(ages) - np.min(ages)
std_dev_age = np.std(ages)
q1 = np.percentile(ages, 25)
q3 = np.percentile(ages, 75)
iqr = q3 - q1

# Display results
print(f"Mean Age: {mean_age:.2f}")
print(f"Median Age: {median_age}")
print(f"Mode Age: {mode_age}")
print(f"Range: {range_age}")
print(f"Standard Deviation: {std_dev_age:.2f}")
print(f"Interquartile Range (IQR): {iqr}")

# Interpretation example
if std_dev_age > 10:
    print("- The data has high variability.")
else:
    print("- The data has low variability.")

# -----------------------------------------------
# PART 2: VISUALIZATION TECHNIQUES - Box Plot and Histogram
# -----------------------------------------------

print("\n=== PART 2: VISUALIZATION TECHNIQUES ===")

plt.figure(figsize=(12, 5))

# Box plot to visualize spread and outliers
plt.subplot(1, 2, 1)
plt.boxplot(ages, vert=True, patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title("Box-and-Whisker Plot of Ages")
plt.ylabel("Age")

# Histogram to show frequency distribution
plt.subplot(1, 2, 2)
plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Ages")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# -----------------------------------------------
# PART 3: KEY STEPS FOR BUILDING DISTRIBUTIONS
# -----------------------------------------------

print("\n=== PART 3: KEY STEPS FOR BUILDING DISTRIBUTIONS ===")

# Step 1: Define bins for age ranges
bins = [15, 25, 35, 45, 55, 65, 75]

# Step 2: Calculate frequencies and relative frequencies
frequency, bin_edges = np.histogram(ages, bins=bins)
relative_frequency = frequency / len(ages)

# Step 3: Create a frequency table
distribution_df = pd.DataFrame({
    "Age Range": [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)],
    "Frequency": frequency,
    "Relative Frequency": relative_frequency.round(3)
})

print("Frequency Distribution Table:")
print(distribution_df)

# Step 4: Visualize frequency polygon with histogram
midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])

plt.figure(figsize=(8, 5))
plt.hist(ages, bins=bins, color='lightgray', edgecolor='black', alpha=0.7, label='Histogram')
plt.plot(midpoints, frequency, color='red', marker='o', linewidth=2, label='Frequency Polygon')
plt.title("Age Distribution with Frequency Polygon")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.xticks(bins)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Interpretation: Skewness
lower_half = np.sum(ages < median_age)
upper_half = np.sum(ages > median_age)

print("\nSkewness Interpretation:")
if lower_half < upper_half:
    print("- The data is LEFT-SKEWED (more older individuals, tail on the left).")
elif lower_half > upper_half:
    print("- The data is RIGHT-SKEWED (more younger individuals, tail on the right).")
else:
    print("- The data is approximately symmetric.")
```

QUESTION: How did adding outliers affect the mean and median?


Introducing an extreme outlier of 1000 ms into the original base network latency dataset (200 values randomly generated between 10 and 199 ms, mean ≈102.37 ms, median 102.0 ms) influenced the mean, but not the median, demonstrating that they are differently sensitive to skewed distributions. The new mean increased to ≈110.49 ms (an increase of about 8 ms, billed as ~8%), because, as noted above, the mean is calculated from the sum of all values, then divided by n; thus, it is influenced by the outlier's high value, impacting the representation of normal latencies, and creating the illusion of more average delay. Meanwhile, the median remains the same, at 102.0 ms, because its value relies simply on value sort position (the 101st in this n=201 dataset), making it robust to extreme values that shift the central ordering without changing the corresponding median position. This is a key reason the median is often preferred when considering right-skewed datasets, like the normal dataset of latencies, where some rare value in the high distribution should not drag down the perceived normal distribution, while the mean is clearly susceptible to asymmetry.

### **Uncertainty in Sample Statistics**

In data analysis, we often rely on samples to estimate characteristics of a larger population. However, because a sample is only a subset of the entire population, there is always uncertainty in how accurately sample results reflect the true population values. This uncertainty must be clearly communicated through statistical measures and visualizations to avoid misleading conclusions. By using tools like confidence intervals and visual aids such as error bars, analysts can provide a clearer picture of the reliability of their findings. These techniques are essential for engineers and data scientists when making decisions based on limited data, such as system monitoring, user behavior analysis, or predictive modeling.

#### Statistical Inference and Sampling
- Statistical inference is the process of using a sample to draw conclusions about a population.
- Uncertainty arises because samples are incomplete representations of the population.
- Common real-world applications include political polling, market research, and quality control.

#### Confidence Intervals (CIs)
- A confidence interval shows a range of values that likely contains the population parameter. It can be showed in two variations:
- Formula:
    - Mean: sample mean ± margin of error
    - Proportion: sample proportion ± margin of error
- Influencing factors:
    - Sample size: Larger samples reduce uncertainty and narrow the interval.
    - Data variability: Greater variability increases uncertainty and widens the interval.
    - Confidence level: Higher confidence levels (e.g., 95% or 99%) require wider intervals.

#### Visualization Techniques
- Error Bars: Visually represent the margin of error around a point estimate.
- Box-and-Whisker Plots: Highlight variability, median, and outliers.
- Violin Plots: Combine box plots with density plots for a richer understanding of distribution shape.
- Starting the vertical axis at zero prevents misleading visuals unless comparing small differences.

#### Practical Examples
- Political Polling: Estimating the proportion of voters supporting a candidate.
- Market Research: Estimating average customer spending.
- Health Analytics: Comparing average lifespans between male and female groups.
- Engineering Context: Detecting network performance issues or system failures through error margins.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# ===============================
# PART 1: Statistical Inference and Sampling
# ===============================
print("=== Part 1: Statistical Inference and Sampling ===")

# Population data: Simulate a population of 10,000 network latency values (in ms)
population = np.random.normal(loc=100, scale=15, size=10000)  # mean=100ms, std=15ms

# Draw a random sample of 50
sample_size = 50
sample = np.random.choice(population, sample_size, replace=False)

# Calculate population and sample means
population_mean = np.mean(population)
sample_mean = np.mean(sample)

print(f"Population Mean (True): {population_mean:.2f} ms")
print(f"Sample Mean (Estimate): {sample_mean:.2f} ms")
print(f"Sample Size: {sample_size}", end="\n\n")

# ===============================
# PART 2: Confidence Intervals
# ===============================
print("=== Part 2: Confidence Intervals ===")

# Calculate 95% confidence interval for the sample mean
confidence_level = 0.95
std_error = np.std(sample, ddof=1) / np.sqrt(sample_size)
margin_of_error = stats.t.ppf((1 + confidence_level) / 2, df=sample_size-1) * std_error

ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

print(f"95% Confidence Interval: [{ci_lower:.2f}, {ci_upper:.2f}] ms")

# ===============================
# PART 3: Visualization Techniques
# ===============================

# Create a dataframe for plotting
population_subset = np.random.choice(population, 200, replace=False)  # 200 for visualization

# Boxplot directly from separate arrays
plt.figure(figsize=(8, 5))
plt.boxplot(
    [population_subset, sample],
    tick_labels=['Population (Subset of 200)', 'Sample (50)'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue')
)
plt.title("Population vs Sample Variability")
plt.ylabel("Latency (ms)")
plt.show()
```

### **Uncertainty in Predictive Models**

Predictive analytics uses past data to create models that forecast future outcomes. These models are widely applied in engineering and business, such as predicting network latency, product sales, or travel times. However, predictions are never exact, and uncertainty must be considered. Visualization of this uncertainty helps engineers and decision-makers understand the reliability of forecasts, improving real-world decision-making.

#### Predictive Models and Point Estimates
- Predictive models generate point estimates, which are single predicted values for future observations.
- These estimates contain error because real-world systems involve randomness and variability.

#### Prediction Intervals
- A prediction interval gives a range where future outcomes are likely to occur with a certain confidence level (e.g., 95%).
- Similar to confidence intervals used in statistical inference, but focused on predicting individual future observations rather than population parameters.

#### Simple Linear Regression Example
- Linear regression models relationships between two variables:
    - Independent variable (x): factor used for prediction (e.g., number of requests in a delivery route).
    - Dependent variable (y): outcome being predicted (e.g., travel time).
- Visualization includes:
    - Historical data points.
    - Regression line showing predicted values.
    - Upper and lower dashed lines showing prediction interval boundaries.
- Key observation: intervals are narrower near the average values of the independent variable and wider at extremes, indicating higher uncertainty for unusual cases.

#### Time Series Models Example
- Used for data measured sequentially over time (e.g., quarterly sales).
- Predictions account for patterns like seasonality or trends.
- Visualization shows:
    - Past data as a continuous line.
    - Future point predictions as a lighter line.
    - Prediction intervals widening as the forecast extends further into the future.
- Important insight: uncertainty grows over time because long-term predictions are inherently less reliable.

```python
# ==============================================
# Uncertainty in Predictive Models
# ==============================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

# ==============================================
# PART 1: Predictive Models and Point Estimates
# ==============================================
print("=== Part 1: Predictive Models and Point Estimates ===")

# Generate random data: relationship between requests and travel time
num_requests = np.random.randint(1, 10, size=10)  # independent variable (random integers 1–9)
travel_time = np.random.normal(loc=3, scale=0.5, size=10)  # dependent variable with random noise

# Fit a simple linear regression model
X = num_requests.reshape(-1, 1)
y = travel_time
model = LinearRegression().fit(X, y)

# Predict point estimates
predictions = model.predict(X)

print("Regression Model Summary:")
print(f"Intercept: {model.intercept_:.3f}, Slope: {model.coef_[0]:.3f}\n")

print("Sample of Predictions:")
for i, p in zip(num_requests, predictions):
    print(f"Requests: {i}, Predicted Travel Time: {p:.2f} hours")

# ==============================================
# PART 2: Prediction Intervals
# ==============================================
print("\n=== Part 2: Prediction Intervals ===")

# Use statsmodels to compute prediction intervals
X_with_const = sm.add_constant(X)  # statsmodels requires intercept column
ols_model = sm.OLS(y, X_with_const).fit()

# Generate prediction intervals
pred_summary = ols_model.get_prediction(X_with_const).summary_frame(alpha=0.05)  # 95% PI
lower_bounds = pred_summary["obs_ci_lower"]
upper_bounds = pred_summary["obs_ci_upper"]

# Display prediction intervals
print(pred_summary[["mean", "obs_ci_lower", "obs_ci_upper"]])

# ==============================================
# PART 3: Visualization for Linear Regression
# ==============================================
print("\n=== Part 3: Visualization for Linear Regression ===")

# Sort data for a cleaner plot
sorted_indices = np.argsort(num_requests)
sorted_requests = num_requests[sorted_indices]
sorted_predictions = predictions[sorted_indices]
sorted_lower = lower_bounds[sorted_indices]
sorted_upper = upper_bounds[sorted_indices]
sorted_actual = travel_time[sorted_indices]

plt.figure(figsize=(8, 6))
plt.scatter(sorted_requests, sorted_actual, color='blue', label='Observed Data')
plt.plot(sorted_requests, sorted_predictions, color='orange', label='Regression Line')
plt.fill_between(sorted_requests, sorted_lower, sorted_upper,
                 color='orange', alpha=0.3, label='95% Prediction Interval')
plt.xlabel("Number of Requests")
plt.ylabel("Travel Time (hours)")
plt.title("Linear Regression with Prediction Intervals")
plt.legend()
plt.show()

# ==============================================
# PART 4: Time Series Forecasting with Prediction Intervals
# ==============================================
print("\n=== Part 4: Time Series Forecasting ===")

# Generate random time series data for quarterly sales
time_periods = 17
quarters = np.arange(1, time_periods + 1)
sales = np.random.normal(loc=150, scale=20, size=time_periods)  # random sales data

# Create dataframe
df_sales = pd.DataFrame({"Quarter": quarters, "Sales": sales})

# Simulate future predictions
future_periods = 8
future_quarters = np.arange(time_periods + 1, time_periods + future_periods + 1)

# Random predictions with added seasonality-like variation
future_sales_pred = np.random.normal(loc=150, scale=15, size=future_periods)

# Add growing uncertainty to future predictions
future_uncertainty = np.linspace(10, 30, future_periods)
lower_future = future_sales_pred - future_uncertainty
upper_future = future_sales_pred + future_uncertainty

# Combine into dataframe
forecast_df = pd.DataFrame({
    "Quarter": future_quarters,
    "Predicted_Sales": future_sales_pred,
    "Lower_PI": lower_future,
    "Upper_PI": upper_future
})

# ==============================================
# Visualization for Time Series
# ==============================================
plt.figure(figsize=(10, 6))
plt.plot(df_sales["Quarter"], df_sales["Sales"], color='darkblue', label='Historical Sales')
plt.plot(forecast_df["Quarter"], forecast_df["Predicted_Sales"], color='lightblue', label='Forecast')
plt.fill_between(forecast_df["Quarter"], forecast_df["Lower_PI"], forecast_df["Upper_PI"],
                 color='lightblue', alpha=0.3, label='95% Prediction Interval')
plt.xlabel("Quarter")
plt.ylabel("Sales ($1000s)")
plt.title("Time Series Forecast with Prediction Intervals")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

# Set random seed for reproducibility
np.random.seed(42)

# Step 1: Create a Synthetic Dataset
print("=== Step 1: Synthetic Dataset Creation ===")

# Generate 200 random ages between 15 and 70 (inclusive)
ages = np.random.randint(15, 71, 200)  # randint(15,71) includes up to 70

print(f"Dataset created: 200 ages ranging from {np.min(ages)} to {np.max(ages)}.")
print(f"Sample ages (first 10): {ages[:10]}")
print("This dataset will be used for all subsequent steps.\n")

# Step 2: Core Concepts - Frequency and Relative Frequency
print("=== Step 2: Frequency and Relative Frequency Table ===")

# Define bins for 10-year age ranges (adjusted for max=70)
bins = [15, 25, 35, 45, 55, 65, 71]  # Covers 15-24, 25-34, ..., 65-70

# Calculate frequency and relative frequency
frequency, bin_edges = np.histogram(ages, bins=bins)
relative_frequency = frequency / len(ages)

# Create age range labels
age_ranges = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]

# Frequency table
distribution_df = pd.DataFrame({
    "Age Range": age_ranges,
    "Frequency": frequency,
    "Relative Frequency": relative_frequency.round(3)
})

print("Frequency Distribution Table:")
print(distribution_df)
print(f"\nTotal Frequency: {np.sum(frequency)} (matches n=200)")
print(f"Total Relative Frequency: {np.sum(relative_frequency):.3f} (sums to 1.000)\n")

# Step 3: Visualization Techniques - Histogram with Frequency Polygon
print("=== Step 3: Histogram and Frequency Polygon ===")

# Calculate midpoints for frequency polygon
midpoints = 0.5 * (bin_edges[1:] + bin_edges[:-1])

# Create plot
plt.figure(figsize=(10, 6))
plt.hist(ages, bins=bins, color='skyblue', edgecolor='black', alpha=0.7, label='Histogram')
plt.plot(midpoints, frequency, color='red', marker='o', linewidth=2, label='Frequency Polygon')
plt.title("Age Distribution: Histogram with Frequency Polygon")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.xticks(bins)
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.show()

print("Visualization Notes:")
print("- Histogram bars show absolute frequencies per bin; peaks indicate most common age groups.")
print("- Frequency polygon connects midpoints, smoothing the distribution to reveal overall shape (e.g., potential skew).")
print("- With these bins, the plot highlights multimodal or uniform trends in the data.\n")

# Step 4: Statistical Measures
print("=== Step 4: Statistical Measures ===")

# Measures of central tendency
mean_age = np.mean(ages)
median_age = np.median(ages)
mode_age = pd.Series(ages).mode()[0]  # First mode if multimodal

# Measures of variability
range_age = np.max(ages) - np.min(ages)
std_dev_age = np.std(ages)
q1 = np.percentile(ages, 25)
q3 = np.percentile(ages, 75)
iqr = q3 - q1

# Display results
print(f"Central Tendency:")
print(f"  Mean Age: {mean_age:.2f}")
print(f"  Median Age: {median_age:.2f}")
print(f"  Mode Age: {mode_age}")

print(f"\nVariability:")
print(f"  Range: {range_age}")
print(f"  Standard Deviation: {std_dev_age:.2f}")
print(f"  Interquartile Range (IQR): {iqr:.2f} (Q1={q1:.2f}, Q3={q3:.2f})")

print("\nInterpretation:")
if mean_age > median_age:
    print("- Slight right skew (mean pulled by higher ages).")
else:
    print("- Approximately symmetric distribution.")
print("- High std dev indicates spread across ages; IQR focuses on middle 50%.\n")

# Step 5: Box Plot and Histogram Comparison
print("=== Step 5: Box Plot and Histogram Comparison ===")

plt.figure(figsize=(12, 5))

# Box-and-whisker plot for spread and outliers
plt.subplot(1, 2, 1)
plt.boxplot(ages, vert=True, patch_artist=True,
            boxprops=dict(facecolor='lightblue'),
            flierprops=dict(marker='o', markerfacecolor='red', markersize=5))
plt.title("Box-and-Whisker Plot of Ages")
plt.ylabel("Age")
plt.grid(axis='y', alpha=0.3)

# Separate histogram for frequency distribution (using same bins)
plt.subplot(1, 2, 2)
plt.hist(ages, bins=bins, color='lightgreen', edgecolor='black', alpha=0.7)
plt.title("Histogram of Age Distribution")
plt.xlabel("Age Range")
plt.ylabel("Frequency")
plt.xticks(bins)

plt.tight_layout()
plt.show()

print("Comparison Notes:")
print("- Box plot summarizes quartiles, median (line in box), and outliers (red dots); shows symmetric spread with possible extremes.")
print("- Histogram reveals detailed frequency shape (e.g., bin heights); complements box by showing density.")
print("- Together, they detect outliers (box) and distribution form (histogram); no major outliers here, but check for skew.\n")

# Step 6: Prediction and Uncertainty Visualization
print("=== Step 6: Simple Linear Regression with Prediction Intervals ===")

# Subset: Random 10 values from ages as x (independent variable)
x_subset = np.random.choice(ages, 10, replace=False)

# y: Random dependent variable with noise (e.g., simulated "experience years" related to age)
y = 0.1 * x_subset + np.random.normal(5, 2, 10)  # Weak linear relation + noise

# Fit linear regression
X = x_subset.reshape(-1, 1)
model = LinearRegression().fit(X, y)
predictions = model.predict(X)

print("Regression Model Summary (n=10 subset):")
print(f"  Intercept: {model.intercept_:.3f}, Slope: {model.coef_[0]:.3f}")
print(f"  R²: {model.score(X, y):.3f} (explains variance; low due to noise)")

# Compute 95% prediction intervals using statsmodels
X_with_const = sm.add_constant(X)
ols_model = sm.OLS(y, X_with_const).fit()
pred_summary = ols_model.get_prediction(X_with_const).summary_frame(alpha=0.05)  # 95% PI
lower_bounds = pred_summary["obs_ci_lower"]
upper_bounds = pred_summary["obs_ci_upper"]

# Sort for plotting
sorted_indices = np.argsort(x_subset)
sorted_x = x_subset[sorted_indices]
sorted_y = y[sorted_indices]
sorted_predictions = predictions[sorted_indices]
sorted_lower = lower_bounds[sorted_indices]
sorted_upper = upper_bounds[sorted_indices]

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(sorted_x, sorted_y, color='green', label='Observed Data', s=50)
plt.plot(sorted_x, sorted_predictions, color='orange', label='Regression Line', linewidth=2)
plt.fill_between(sorted_x, sorted_lower, sorted_upper, color='orange', alpha=0.3, label='95% Prediction Interval')
plt.xlabel("Age (x)")
plt.ylabel("Dependent Variable (y, e.g., Experience Years)")
plt.title("Linear Regression on Age Subset with 95% Prediction Intervals")
plt.legend()
plt.grid(alpha=0.3)
plt.show()

"""

"""
```
